```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("newclean_data.csv")

dataset.head()
```

```python
dataset.columns
```

```python
dataset["2022.0"] = pd.to_numeric(dataset["2022.0"], errors="coerce")
dataset_cleaned = dataset.dropna(subset=["2022.0"])
```

```python
top_10 = dataset_cleaned.sort_values(by="2022.0", ascending=False).head(10)
```

```python
plt.figure(figsize=(14, 7))
bars = plt.bar(top_10["Country Name"], top_10["2022.0"], color='mediumseagreen', edgecolor='black')

# Add grid, labels, and values on top of each bar
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Top 10 Most Populous Countries in 2022", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=13)
plt.ylabel("Population", fontsize=13)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height):,}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
